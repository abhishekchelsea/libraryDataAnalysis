```python
bk = pd.read_csv(
    "/home/aca/Documents/S6_Lab/Projects/libraryDataAnalysis/pattern_mining/unq_bks.csv", index_col=0)
# index will be used as unique id
```

```python
# This operation took 2m 25 seconds !!
lst_of_idxs = []
for index, row in df.iterrows():
    _idx = bk.index[(bk['title'] == row['title']) & (bk['publisher'] == row['publisher']) & (bk['author'] == row['author'])].tolist()
    if (len(_idx) != 1):
        lst_of_idxs.append(np.nan)
    else:
        lst_of_idxs.append(_idx[0])
df['book_id'] = lst_of_idxs
df
# 41088 rows
```

```python
# drop rows with book_id = null. 
df2 = df.dropna(subset=['book_id'])
df2
# 789 rows are eliminated. Now we have 40329 rows 
```

```python
# Write data into csv. 
df2 = df2.astype(
    {
        "book_id": "int"
    })
df2.to_csv("book_issue_with_book_id.csv", index=False)
```
