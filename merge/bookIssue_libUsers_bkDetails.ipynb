```python
import pandas as pd
```

```python
lib_user = pd.read_csv(
    "/home/aca/Documents/S6_Lab/Projects/libraryDataAnalysis/cleaning/outputs/clean6_library_users.csv")
bk_issue = pd.read_csv(
    "/home/aca/Documents/S6_Lab/Projects/libraryDataAnalysis/cleaning/outputs/clean2_bookIssue.csv")
bk_details = pd.read_csv(
    "/home/aca/Documents/S6_Lab/Projects/libraryDataAnalysis/cleaning/outputs/clean1_bookDetails.csv")


# rename adm_no column to id for easy merging
bk_issue = bk_issue.rename({"adm_no": "id"}, axis=1)

# lib_user  # 2924 rows #id
# bk_issue # 41145 rows #id #acc_no
# bk_details #
```

```python
# merge book details with book issue details on acc_no
bk_issue_bk_detials = pd.merge(bk_issue, bk_details, on="acc_no")

# Some rows are missing probably due to mismatch
# bk_issue_bk_detials.to_csv("bk_issue_bk_details.csv", index=False)
```

```python
# merge lib user details with book issue details on id
bk_issue_lib_user = pd.merge(bk_issue, lib_user, on="id")

# bk_issue_lib_user.to_csv("bk_issue_lib_user.csv", index=False)
```

```python
# Merge bk issue, bk details and lib user

bkIssue_libUser_bkDetails = pd.merge(bk_issue_bk_detials, lib_user, on="id")
# bkIssue_libUser_bkDetails
```

```python
# checking if title_x and title_y are the same
for index, row in bkIssue_libUser_bkDetails.iterrows():
    if (row["title_x"] != row["title_y"]):
        raise Exception
```

```python
# Drop redundant columns

# drop title_y
bkIssue_libUser_bkDetails.drop("title_y", axis=1, inplace=True)
```

```python
bkIssue_libUser_bkDetails.rename(
    {
        "title_x": "title",
        "Sl No": "sl_no"
        }, axis=1, inplace=True)
bkIssue_libUser_bkDetails
```

```python
# Convert float to int

# convert sl_no 
bkIssue_libUser_bkDetails = bkIssue_libUser_bkDetails.astype(
    {
        "sl_no": "int"
        })

# convert batch. 
# NOTE: NaN values will be replaced by 0 
bkIssue_libUser_bkDetails["batch"] = bkIssue_libUser_bkDetails["batch"].fillna(0).astype(int)

# bkIssue_libUser_bkDetails.dtypes
# bkIssue_libUser_bkDetails
```

```python
bkIssue_libUser_bkDetails
```

```python
# Write to csv
bkIssue_libUser_bkDetails.to_csv("temp.csv", index=False)
```
